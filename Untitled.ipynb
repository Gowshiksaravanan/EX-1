```python
import statistics as st
import pandas as pd
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
da=pd.read_csv("Z:\data analytics\mine.csv")
df=pd.DataFrame(da)
print(" Arithmetic Mean of age")
print(st.mean(df.Age))
print(" Arithmetic Mean of Potential")
print(st.mean(df.Potential))
print(" Arithmetic Mean of Overall")
print(st.mean(df.Overall))
```

```python
print(" Harmonic Mean of age")
print(st.harmonic_mean(df.Age))
print(" Harmonic Mean of Potential")
print(st.harmonic_mean(df.Potential))
print(" Harmonic Mean of Overall")
print(st.harmonic_mean(df.Overall))
```

```python
print(" Geometric Mean of Overall")
print(stats.gmean(df.Overall))
print(" Geometric Mean of age")
print(stats.gmean(df.Age))
print(" Geometric Mean of Potential")
print(stats.gmean(df.Potential))
```

```python
print("Interquartile range of Age")
print(stats.iqr(df.Age))
print("Interquartile range of Overall")
print(stats.iqr(df.Overall))
print("Interquartile range of Potential")
print(stats.iqr(df.Potential))
```

```python
print("Boxplot")
dn = pd.DataFrame(np.random.rand(100, 5), columns=['Age', 'Overall', 'Potential', 'Special', 'Jersey'])
dn.plot.box(grid='True')
```

```python
print("Histogram")
df.Overall.hist()
plt.plot()
plt.xlabel("value")
plt.ylabel("No.of players")
```

```python
print("Bar Graph")
ax=df['Position'].value_counts()
index = np.arange(len(ax))
plt.figure(figsize=(50,15))
plt.bar(index, ax)
plt.xlabel("Position",fontsize=20)
plt.ylabel("Count",fontsize=20)
plt.xticks(index, df["Position"].unique(),fontsize=20)
plt.yticks(fontsize=20)
```

```python
print("Pie chart for position")
dx=df[:20]
ax=dx['Position'].value_counts()
plt.pie(ax.tolist(),labels=list(dx["Position"].unique()),startangle=90,autopct='%1.1f%%')
plt.show()
```
